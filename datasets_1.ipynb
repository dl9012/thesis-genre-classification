```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import json

data = pd.read_csv("movie.metadata.tsv", sep="\t", header=None)
data.head()

```

```python
data.columns = ["movie_id",1,"movie_name","year",4,5,6,7,"genre"]
data["year"] = data["year"].astype(str).str[:4]


data = data.drop(columns=[1,4,5,6,7])
data.head()
```

```python
# an empty list
languages = [] 

# extract languages from column 6
for i in data.iloc[:, 6]: 
  languages.append(list(json.loads(i).values())) 

# add to 'movies' dataframe
data['genre_new'] = languages

#data = data.drop(columns=["genre"])
data.head()
```

```python
data.shape
```

```python
# drop every row where genre_new does not contain English Language
data = data[data['genre_new'].apply(lambda x: 'English Language' in x)]
data.head()
```

```python
data.shape
```

```python
# create dataframe from plot_summaries.txt
plot_summaries = pd.read_csv("plot_summaries.txt", sep="\t", header=None)
plot_summaries.columns = ["movie_id", "plot"]
plot_summaries.head()

# Convert the 'movie_id' column in both dataframes to int
data['movie_id'] = data['movie_id'].astype(int)
plot_summaries['movie_id'] = plot_summaries['movie_id'].astype(int)

# Now merge the dataframes on the 'movie_id' column
data = pd.merge(data, plot_summaries, on="movie_id")
data.head()
```

```python
# an empty list
genres = [] 

# extract genres
for i in data['genre']: 
  genres.append(list(json.loads(i).values())) 

# add to 'movies' dataframe
data['genre_new1'] = genres

data = data.drop(columns=["genre"])
data.head()
```

```python
# check for missing values in the dataframe
movies = data[~(data['genre_new1'].str.len() == 0)]
movies.shape, data.shape
```

### Splitting genres into equally big parts
4 genres, thriller, horror, comedy drama

```python
movies.head()
```

```python
# create a single dataframe with 1500 movies containing the comedy genre, 1500 movies containing the horror genre, 1500 movies containing the thriller genre and 1500 movies containing the drama genre
# if the movie is already in any dataframe, it should not be added again
```

```python
movies2 = movies
# create a single dataframe with 1500 movies containing the comedy genre
# remove all the movies added to the comedy dataframe from the original dataframe

# HÄMTA DE MED MINDRE ANTAL FILMER FÖRST kanske

comedy = movies2[movies2['genre_new1'].apply(lambda x: 'Comedy' in x and not any(genre in x for genre in ['Horror', 'Thriller', 'Drama']))]
comedy = comedy.head(1500)
movies2 = movies2.drop(comedy.index) 

horror = movies2[movies2['genre_new1'].apply(lambda x: 'Horror' in x and not any(genre in x for genre in ['Comedy', 'Thriller', 'Drama']))]
horror = horror.head(1500)
movies2 = movies2.drop(horror.index)

thriller = movies2[movies2['genre_new1'].apply(lambda x: 'Thriller' in x and not any(genre in x for genre in ['Horror', 'Comedy', 'Drama']))]
thriller = thriller.head(1500)
movies2 = movies2.drop(thriller.index)

drama = movies2[movies2['genre_new1'].apply(lambda x: 'Drama' in x and not any(genre in x for genre in ['Horror', 'Thriller', 'Comedy']))]
drama = drama.head(1500)
movies2 = movies2.drop(drama.index)

# combine the dataframes
movies3 = pd.concat([comedy, horror, thriller, drama])
movies3.shape
```

```python
# keep only 'Thriller', 'Horror', 'Comedy' and 'Drama' genres
movies3['genre_new1'] = movies3['genre_new1'].apply(lambda x: [i for i in x if i in ['Thriller', 'Horror', 'Comedy', 'Drama']])
movies3 = movies3[movies3['genre_new1'].str.len() != 0]
movies3.head()
```

```python
#save movies3 without the 'plot' column to a csv file in current folder
movies4 = movies3.drop(columns=["plot"])
movies4.to_csv("movies4.csv", index=False)
```

```python
api_example_list = []

# get the data from the api for the 10 first movies in example_list
# if {"Response":"False","Error":"Movie not found!"} is printed, the movie was not found

for movie, year in example_list:
    url = f"http://www.omdbapi.com/?t={movie}&y={year}&plot=full&apikey={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        movie_data = response.json()
        if movie_data["Response"] != "False":
            api_example_list.append(movie_data)
    else:
        print(f"Error: {response.status_code} - {response.text}")

for movie in api_example_list:
    print(movie)
```
