```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

data = pd.read_csv("movie.metadata.tsv", sep="\t", header=None)
data.head()

```

```python
data.columns = ["movie_id",1,"movie_name","year",4,5,6,7,"genre"]
data["year"] = data["year"].astype(str).str[:4]


#data = data.drop(columns=[1,4,5,6,7])
data.head()
```

```python
import json

# an empty list
languages = [] 

# extract languages from column 6
for i in data.iloc[:, 6]: 
  languages.append(list(json.loads(i).values())) 

# add to 'movies' dataframe
data['genre_new'] = languages

#data = data.drop(columns=["genre"])
data.head()
```

```python
data.shape
```

```python
# drop every row where genre_new does not contain English Language
data = data[data['genre_new'].apply(lambda x: 'English Language' in x)]
data.head()
```

```python
data.shape
```

```python
# create dataframe from plot_summaries.txt
plot_summaries = pd.read_csv("plot_summaries.txt", sep="\t", header=None)
plot_summaries.columns = ["movie_id", "plot"]
plot_summaries.head()

# Convert the 'movie_id' column in both dataframes to int
data['movie_id'] = data['movie_id'].astype(int)
plot_summaries['movie_id'] = plot_summaries['movie_id'].astype(int)

# Now merge the dataframes on the 'movie_id' column
data = pd.merge(data, plot_summaries, on="movie_id")
data.head()
```

```python
import json

# an empty list
genres = [] 

# extract genres
for i in data['genre']: 
  genres.append(list(json.loads(i).values())) 

# add to 'movies' dataframe
data['genre_new1'] = genres

data = data.drop(columns=["genre"])
data.head()
```

```python
# check for missing values in the dataframe
movies = data[~(data['genre_new'].str.len() == 0)]
movies.shape, data.shape
```

```python
# a list containing key value pairs of the 10 first movies in the dataframe and their year
example_list = []
for i in range(10):
    example_list.append((movies.iloc[i]["movie_name"], movies.iloc[i]["year"]))
print(example_list)
```

```python
api_example_list = []

# get the data from the api for the 10 first movies in example_list
# if {"Response":"False","Error":"Movie not found!"} is printed, the movie was not found

for movie, year in example_list:
    url = f"http://www.omdbapi.com/?t={movie}&y={year}&plot=full&apikey={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        movie_data = response.json()
        if movie_data["Response"] != "False":
            api_example_list.append(movie_data)
    else:
        print(f"Error: {response.status_code} - {response.text}")

for movie in api_example_list:
    print(movie)
```

```python
api_example_df = pd.DataFrame(api_example_list)
api_example_df = api_example_df[["Title", "Year", "Genre", "Plot"]]
api_example_df.head()
```
